```python
import pandas as pd
dataset = pd.read_csv('dataset_for_SFYS.csv', index_col=['Unnamed: 0'])
dataset.head()
```

```python
dataset.columns
```

```python
dataset = dataset.drop('Full_name', axis=1)
dataset
```

```python
dataset.Equipments.unique()
```

```python
def fill(row):
    if row['Equipments'] == '0 устройства нет':
        row['Equipments'] = 'Нет'
    else:
        row['Equipments'] = 'Есть'
    return row

dataset = dataset.apply(fill, axis=1)
dataset
```

```python
def fill(row):
    if row['Sex'] == 'женский':
        row['Sex'] = 'female'
    else:
        row['Sex'] = 'male'
    return row

dataset = dataset.apply(fill, axis=1)
dataset
```

```python
def fill(row):
    if row['Equipments'] == 'Нет':
        row['Equipments'] = 'No'
    else:
        row['Equipments'] = 'Yes'
    if row['Status'] == 'Нет':
        row['Status'] = 'No'
    else:
        row['Status'] = 'Yes'
    return row

dataset = dataset.apply(fill, axis=1)
dataset
```

```python
dataset.columns = ['Clinical_record', 'Indexing', 'Sex', 'Age', 'Time_of_hospitalization',
       'Time_of_discharge_from_hospital', 'Diagnosis',
       'Time_of_examination', 'Score', 'Consciousness_rate', 'SpO2',
       'Respiration_rate', 'Systolic_BP', 'Temperature', 'Heart_rate',
       'Equipments', 'Status', 'Days_in_hospital']
dataset
```

```python
dataset.to_csv('dataset_for_SFYS.csv')
```
